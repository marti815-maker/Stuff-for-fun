```python
import glob
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2gray,label2rgb
from skimage.transform import resize
from skimage.segmentation import slic
from skimage.filters import gaussian,meijering, threshold_otsu
from skimage.morphology import binary_closing,disk, remove_small_objects
from scipy.ndimage import binary_fill_holes
```

```python
files=glob.glob('./images/*')
imgs_small=[]
imgs_bw=[]

images=[plt.imread(file) for file in files]

for image in images:
    imgs_small.append(resize(image,(image.shape[0]//5,image.shape[1]//5)))
    imgs_bw.append(rgb2gray(image))

plt.imshow(images[0])
```

```python
def get_mask(image):
    result=meijering(image)
    result=gaussian(result,sigma=5)
    thresh=threshold_otsu(result)
    result=np.where(result>thresh,1,0)
    result=binary_closing(result,disk(20))
    result=binary_fill_holes(result)
    img_size=img.shape[0]*img.shape[1]
    result=remove_small_objects(result,min_size=img_size//40)
    return result
```

```python
for i,img in enumerate(imgs_bw):
    mask=get_mask(img)
    mask_color=np.zeros_like(images[i])
    mask_color[:,:,0]=mask
    mask_color[:,:,1]=mask
    mask_color[:,:,2]=mask
    img=np.where(mask_color==1,images[i],0)
    plt.imsave(f'./masked/masked_{i}.jpeg',img)
    plt.imshow(img)
    plt.show()
```
